```python
import pickle
import pandas as pd
from matplotlib import pyplot as plt
import os
import numpy as np
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
ias = pickle.load(open("inputdays&span", 'rb'))
```

```python
ias
```

```python
for i in range(len(ias)) :
    ias[i] = (ias[i][0], ias[i][1], np.mean(ias[i][2]))
```

```python
ias
```

```python
dates_set = []
for date in pd.date_range("2015-01-01","2021-04-01",freq="A").tolist() :
    dates_set.append(str(date).split(" ")[0])
```

```python
X = np.arange(1,30)
Y = np.arange(1,30)
X,Y = np.meshgrid(X,Y)
Z = np.array([i[2] for i in ias])
Z = Z.reshape(29,29)
```

```python
ax = plt.figure(figsize=(20,10))
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z, 50, cmap='coolwarm')
ax.set_xlabel('input_days')
ax.set_ylabel('span')
ax.set_zlabel('MAE')
```
