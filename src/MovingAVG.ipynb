```python
import pickle
import pandas as pd
from matplotlib import pyplot as plt
import os
import numpy as np
```

```python
ma = pickle.load(open("movingAVG", 'rb'))
```

```python
dates_set = []
for date in pd.date_range("2015-01-01","2021-04-01",freq="QS").tolist() :
    dates_set.append(str(date).split(" ")[0])
```

```python
dates_set
```

```python
plt.figure(figsize=(20,10))
plt.rc('font', size=30)

plt.plot(dates_set, ma['sma'], marker="o", label="sma" )
plt.plot(dates_set, ma['wma'], marker="o", label="wma" )
plt.plot(dates_set, ma['ema'], marker="o", label="ema" )
plt.xticks('')
plt.legend()
```
